```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = ["Humans", "Chickens", "Krill", "Cows", "Ants", "Bacteria"]
y = [350, 40, 150, 520, 3000, 1000000]
plt.style.use("ggplot")
plt.yscale('log')
plt.bar(x, y)
plt.title("Biomass by Population type")
plt.ylabel("Biomass (million tons)")
plt.xlabel("Population type")
plt.savefig("species_population.png", bbox_inches='tight')
```
